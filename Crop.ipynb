```python
import torch
from yolov5.models.yolo import Model
import os
```

```python
model_path = "models/plate_detector.pt"
img_dir = "data/raw/"
crop_dir = "data/normal"
device = torch.device('cpu')
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)
model = model.to(device)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import uuid

model.conf = 0.85
padding = 30
```

```python
files = os.listdir(img_dir)

for image_file in files:
    print(image_file)
    img_path = os.path.join(img_dir, image_file)
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = model(img_rgb)

    boxes = result.xyxy[0]

    for box in boxes:
        x1, y1, x2, y2 = box[:4].cpu().numpy()
        x1 = max(0, int(x1) - padding)
        y1 = max(0, int(y1) - padding)
        x2 = min(img.shape[1], int(x2) + padding)
        y2 = min(img.shape[0], int(y2) + padding)

        # Draw padded bounding box
        cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (255, 0, 0), 10)
        cropped_img = img[y1:y2, x1:x2]

        # Save cropped image
        crop_filename = f'{crop_dir}/{uuid.uuid1()}.jpg'
        cv2.imwrite(crop_filename, cropped_img)
        print(f'Saved cropped plate: {crop_filename}')

    plt.figure(figsize=(5, 5))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show();
```
