```python
import cv2
import numpy as np
import os
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def apply_low_light(img, brightness=0.3):
    low_light_img = np.clip(img * brightness, 0, 255).astype(np.uint8)
    return low_light_img
```

```python
in_dir = os.path.join('data', 'normal')
out_dir = os.path.join('data', 'low_light')
```

```python
brightness = random.uniform(0.05, 0.25)
img = cv2.imread(os.path.join(in_dir, '2.jpg'))

low_light_img = apply_low_light(img, brightness=brightness)
print(brightness)

plt.imshow(cv2.cvtColor(low_light_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
files = os.listdir(in_dir)
```

```python
for image_name in files:
    img = cv2.imread(os.path.join(in_dir, image_name))
    brightness = random.uniform(0.05, 0.25)
    low_light_img = apply_low_light(img, brightness=brightness)
    print(f'Brightness: {brightness}')

    filename = f'{out_dir}/{image_name}'
    cv2.imwrite(filename, low_light_img)
    print(f'Saved low light plate: {filename}')

    plt.imshow(cv2.cvtColor(low_light_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
